# (b)Split data into trainning and testing set

```python
from sklearn.model_selection import train_test_split
X = OJ.drop('Purchase', axis=1)
y = OJ['Purchase']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape)
```

```python
#Encode Categorical Variable "Purchase"
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)
```

```python
# Encode Categorical Variable 'Store7'
X_train_encoded = pd.get_dummies(X_train, columns=['Store7'], drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=['Store7'], drop_first=True)
```

# (c) Fit a support vector classifier with C=0.01


```python
#Import the SVC class from sklearn.svm
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
svm_model = SVC(C=.01, kernel='linear')
svm_model.fit(X_train_scaled, y_train)
```

```python
# Count support vectors
num_support_vectors = len(svm_model.support_vectors_)
num_support_vectors
```

# (d) Calculate training and test error rates

```python
train_error = 1 - svm_model.score(X_train_scaled, y_train)
test_error = 1 - svm_model.score(X_test_scaled, y_test)
print(train_error)
print(test_error)
```

# (e) Use cross-validation to select an optimal C

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [0.01, 0.1, 1, 10]}
grid_search = GridSearchCV(SVC(kernel='linear'), param_grid, cv=5, refit=False)
grid_search.fit(X_train_scaled, y_train)

print("Grid Search Results:")
print("C values:", param_grid['C'])
print("Mean cross-validated score of the best_estimator:", grid_search.best_score_)
print("Best C value:", grid_search.best_params_['C'])
```

# (f)  Retrain a model with the best new value of C and compute the training and test error rates.
Keep the maximum iterations to 100000, max_iter= 100000

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

best_c_value = grid_search.best_params_['C']

best_model = SVC(C=best_c_value, kernel='linear', max_iter=100000)
best_model.fit(X_train_scaled, y_train)

train_predictions = best_model.predict(X_train_scaled)
test_predictions = best_model.predict(X_test_scaled)

train_error_rate = 1 - accuracy_score(y_train, train_predictions)
test_error_rate = 1 - accuracy_score(y_test, test_predictions)

print(f"Training Error Rate with C={best_c_value}: {train_error_rate:.4f}")
print(f"Test Error Rate with C={best_c_value}: {test_error_rate:.4f}")
```

# (g) Repeat parts (b) through (e) using a support vector machine with a radial kernel. Use the default value for gamma.

```python
#Support Vector Machine with Radial Kernel
svm_radial_model = SVC(kernel='rbf')
svm_radial_model.fit(X_train_scaled, y_train)

# Count support vectors
num_support_vectors_radial = len(svm_radial_model.support_vectors_)
print(f"Number of support vectors (radial kernel): {num_support_vectors_radial}")

# Calculate training and test error rates
train_error_radial = 1 - svm_radial_model.score(X_train_scaled, y_train)
test_error_radial = 1 - svm_radial_model.score(X_test_scaled, y_test)

print(f"Training error rate (radial kernel): {train_error_radial:.4f}")
print(f"Test error rate (radial kernel): {test_error_radial:.4f}")
```

# (g) Repeat parts (b) through (e) using a support vector machine with a polynomial kernel. Set degree = 2.

```python
# (h) Support Vector Machine with Polynomial Kernel
svm_poly_model = SVC(kernel='poly', degree=2)
svm_poly_model.fit(X_train_scaled, y_train)

# Count support vectors
num_support_vectors_poly = len(svm_poly_model.support_vectors_)
print(f"Number of support vectors (polynomial kernel): {num_support_vectors_poly}")

# Calculate training and test error rates
train_error_poly = 1 - svm_poly_model.score(X_train_scaled, y_train)
test_error_poly = 1 - svm_poly_model.score(X_test_scaled, y_test)

print(f"Training error rate (polynomial kernel): {train_error_poly:.4f}")
print(f"Test error rate (polynomial kernel): {test_error_poly:.4f}")
```

# (h) Overall, which approach seems to give the best results on this data?

(Ans) Based on the provided results, the radial kernel is preferable for this problem.
The radial kernel model achieves a good balance between fitting the training data well and generalizing to new, unseen data.The lower number of support vectors and lower error rates on both training and test sets suggest that the radial kernel is capturing the underlying patterns in the data more effectively.
